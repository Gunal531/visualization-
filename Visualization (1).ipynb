```python
import numpy as np
import pandas as pd

# Generate sample data
np.random.seed(42)
data = np.concatenate([
    np.random.normal(100, 20, 1000),  # Normal distribution
    np.random.normal(200, 30, 20)     # Some outliers
])

# Calculate measures of level
mean = np.mean(data)
median = np.median(data)
mode = pd.Series(data).mode().values[0]  # Using pandas for mode calculation

# Calculate measures of spread
data_range = np.max(data) - np.min(data)
q1, q3 = np.percentile(data, [25, 75])
iqr = q3 - q1
variance = np.var(data)
std_dev = np.std(data)
coef_var = (std_dev / mean) * 100

# Print results
print(f"Measures of Level:")
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode:.2f}")
print(f"\nMeasures of Spread:")
print(f"Range: {data_range:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Coefficient of Variation: {coef_var:.2f}%")

# Visualize the data
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(data, bins=50, edgecolor='black')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean ({mean:.2f})')
plt.axvline(median, color='green', linestyle='dashed', linewidth=2, label=f'Median ({median:.2f})')
plt.axvline(mode, color='blue', linestyle='dashed', linewidth=2, label=f'Mode ({mode:.2f})')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution of Data with Measures of Central Tendency')
plt.legend()
plt.show()
```
